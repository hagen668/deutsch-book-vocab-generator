```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# print(soup.title) # title w HTML
# print(soup.title.string) # title as string (no HTML)
# print(soup.get_text()) all text (no HTML)
```

```python
data = [element.text for element in soup.find_all("p")]
bookstr = ' '.join(data)
```

```python
tokens = nltk.word_tokenize(bookstr)
text = nltk.Text(tokens)
```

```python
nonPunct = re.compile('.*[A-Za-z].*')
raw_words = [w for w in text if nonPunct.match(w)]
no_stop_words = [w for w in raw_words if not w in stopwords.words("english")]
```

```python
nltk.pos_tag(raw_words)
```
